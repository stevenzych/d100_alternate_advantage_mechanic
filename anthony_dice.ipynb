Plot percentage chance to get below any given skill level for 100 rolls.

```python
# Simulates trying to role under a skill level in Mythras d100 system, with Anthonyvantage.
skill_dict = {}
for x in range(100):
    skill_level = x
    skill_list = []
    for i in range(100000):
        a, b = rn.randint(0, 10, 2)
        first_combo = (int(str(a)+str(b)))
        second_combo = (int(str(b)+str(a)))
        if first_combo < second_combo:
            winner = first_combo
        else:
            second_combo = winner
        skill_list.append(winner < skill_level)
    skill_dict[x] = np.average(skill_list)
```

```python
# Simulates trying to role under a skill level in Mythras d100 system, with Sadthonysauce.
fail_dict = {}
for x in range(100):
    skill_level = x
    skill_list = []
    for i in range(100000):
        a, b = rn.randint(0, 10, 2)
        first_combo = (int(str(a)+str(b)))
        second_combo = (int(str(b)+str(a)))
        if first_combo > second_combo:
            winner = first_combo
        else:
            second_combo = winner
        skill_list.append(winner < skill_level)
    fail_dict[x] = np.average(skill_list)
```

```python
sns.set_style(style='whitegrid')
plt.figure(figsize=(12,8))
plt.scatter(skill_dict.keys(), skill_dict.values(), marker='+')
plt.scatter(fail_dict.keys(), fail_dict.values(), marker='+');
```
