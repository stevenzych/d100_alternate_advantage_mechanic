# Introduction

This notebook looks at different advantage and disadvantage mechanics for d100 RPG systems. Ultimately, we look at the effects of the following game mechanic:
- Roll 10-sided dice separately. Treat zeroes as `0` as opposed to `10`.
- Create two string options with the outputs, such that a roll like "1 and 9" can either be `19` or `91`.
- Select either the lower or higher combination based on whether you have advantage or disadvantage on the roll.

The effects of this rolling mechanic are looked at both in terms of distribution and rolling-below-X-percent on a hypothetical in-game skill check.

> *Rolling in this style will be referred to as "Tony dice," after my friend whose idea it was.*

# Packages

```python
import numpy as np
import numpy.random as rn
rn.seed(0)

import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style(style='whitegrid')

import pandas as pd
pd.set_option('max_rows', None)
```

# Rolling `2d6 Tony`

## Basic Distribution

The first example follows the above protocol. But instead with 6-sided dice.

```python
roll_2d6 = []
for i in range(1000000):
    a = rn.randint(1, 7)
    b = rn.randint(1, 7)
    roll_2d6.append(int(str(a)+str(b)))
    roll_2d6.append(int(str(b)+str(a)))
```

```python
plt.figure(figsize=(10,6))
sns.distplot(roll_2d6, bins='auto', color='b')
plt.title('Basic Distribution Of 2d6 Tony')
plt.xlabel('Roll')
plt.ylabel('Likelihood');
```

## Adding (Dis)advantage To Rolls

Now we're doing the same thing, but adding in the advantage and disadvantage.

### Advantage

```python
lower_list = []
for i in range(100000):
    a = rn.randint(1, 7)
    b = rn.randint(1, 7)
    first_combo = (int(str(a)+str(b)))
    second_combo = (int(str(b)+str(a)))
    if first_combo < second_combo:
        lower_list.append(first_combo)
    else:
        lower_list.append(second_combo)
```

```python
plt.figure(figsize=(10,6))
sns.distplot(lower_list, bins='auto', color='green')
plt.title('Distribution Of 2d6 Tony With Advantage')
plt.xlabel('Roll')
plt.ylabel('Likelihood');
```

### Disadvantage

```python
upper_list = []
for i in range(100000):
    a = rn.randint(1, 7)
    b = rn.randint(1, 7)
    first_combo = (int(str(a)+str(b)))
    second_combo = (int(str(b)+str(a)))
    if first_combo > second_combo:
        upper_list.append(first_combo)
    else:
        upper_list.append(second_combo)
```

```python
plt.figure(figsize=(10,6))
sns.distplot(upper_list, color='orange', bins='auto')
plt.title('Distribution Of 2d6 Tony With Disadvantage')
plt.xlabel('Roll')
plt.ylabel('Likelihood');
```

### Both Distributions

```python
plt.figure(figsize = (10,6))
sns.kdeplot(lower_list, color='green')
sns.kdeplot(upper_list, color='orange')
plt.title('(Dis)advantaged 2d6 Tony Distributions')
plt.xlabel('Roll')
plt.ylabel('Likelihood');
```

## Likelihood To Roll Under Skill Threshold For Either Condition

Basically, the following code simulates trying to roll `2d10 Tony` *under* a certain skill with either advantage or disadvantage. In the Mythras game system, for example, if a player has a Medicine skill of `45`, they must roll *below* `45` to successfully perform a Medicine check, and so on. The following plots answer the question:
> "Based on my skill level in this class, how likely am I to succeed on a check when rolling Tony?"

### Advantage

```python
adv_dict = {}
for x in range(100):
    skill_level = x
    skill_list = []
    for i in range(100000):
        a, b = rn.randint(0, 10, 2)
        first_combo = (int(str(a)+str(b)))
        second_combo = (int(str(b)+str(a)))
        if first_combo < second_combo:
            winner = first_combo
        else:
            winner = second_combo
        skill_list.append(winner < skill_level)
    adv_dict[x] = np.average(skill_list)
```

### Disadvantage

```python
disadv_dict = {}
for x in range(100):
    skill_level = x
    skill_list = []
    for i in range(100000):
        a, b = rn.randint(0, 10, 2)
        first_combo = (int(str(a)+str(b)))
        second_combo = (int(str(b)+str(a)))
        if first_combo > second_combo:
            winner = first_combo
        else:
            winner = second_combo
        skill_list.append(winner < skill_level)
    disadv_dict[x] = np.average(skill_list)
```

### Both Distributions

```python
plt.figure(figsize=(10,5))
plt.scatter(adv_dict.keys(), adv_dict.values(), marker='+', c='green')
plt.scatter(disadv_dict.keys(), disadv_dict.values(), marker='+', c='orange')
plt.title('Likelihood To Succeed On A Skill Check When Rolling 2d10 Tony')
plt.xlabel('Skill Threshhold')
plt.ylabel('Likelihood Of Success');
```

### Percentage Chance To Roll Under A Skill Check

These are just the dictionaries that show either condition's likelihoods to beat a skill check by rolling under said value. Please keep in mind that these values are not perfect as they were computed by simulating dice rolls, and convergence may not have occurred.

```python
df = pd.DataFrame.from_dict(adv_dict, orient='index', columns=['Adv'])
df['Adv'] = df['Adv']*100
df['Disadv'] = [x*100 for x in disadv_dict.values()]
df.head(100)
```
